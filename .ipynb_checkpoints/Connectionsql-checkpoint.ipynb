```python
import pyodbc

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost;"
    "DATABASE=SupplyChainDB;"
    "Trusted_Connection=yes;"
)

print("✅ Connected successfully!")
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mssql+pyodbc://localhost/SupplyChainDB?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server"
)
```

```python
import pandas as pd

df = pd.read_csv("cleaned_supply_chain.csv")
```

```python
df.to_sql('supply_chain_data' ,
          con=engine,
          if_exists='replace' ,
          index=False )
Print('Successful')
```
