```python
import pandas as pd 
import numpy as np
```

```python
df= pd.read_csv('DataCoSupplyChainDataset.csv' , encoding='Latin1')
```

```python
df.head(5)
```

```python
df.columns
```

```python
import numpy as np

cols_missing = ["Customer City", "Customer State", "Sales"]

for col in cols_missing:
    df.loc[df.sample(frac=0.05).index, col] = np.nan
```

```python
df["Sales"] = df["Sales"].astype(str)
df["Order Item Quantity"] = df["Order Item Quantity"].astype(str)
```

```python
cols_inconsistent = ["Customer Country", "Customer State"]

for col in cols_inconsistent:
    df.loc[df.sample(frac=0.05).index, col] = "IND"
    df.loc[df.sample(frac=0.05).index, col] = "India "
```

```python
cols_case = ["Order Status", "Shipping Mode"]

for col in cols_case:
    df.loc[df.sample(frac=0.05).index, col] = "delivered"
    df.loc[df.sample(frac=0.05).index, col] = "DELIVERED"
```

```python
df = pd.concat([df, df.sample(200)], ignore_index=True)
```

```python
df.loc[df.sample(frac=0.02).index, "Sales"] = 999999
df.loc[df.sample(frac=0.02).index, "Order Item Quantity"] = 999
```

```python
df.loc[df.sample(frac=0.03).index, "Sales"] = -500
df.loc[df.sample(frac=0.03).index, "Order Item Quantity"] = -10
```

```python
date_cols = ["order date (DateOrders)", "shipping date (DateOrders)"]

for col in date_cols:
    df[col] = df[col].astype(str)
```

```python
df.to_csv("SupplyChainData.csv", index=False)
```
