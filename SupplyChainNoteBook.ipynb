Changing DataTypes 

```python
df['Sales'].dtypes
```

```python

df["Sales"] = pd.to_numeric(df["Sales"] , errors='coerce')
```

```python
df['Sales'].dtypes
```

```python
df['Order Item Quantity'].dtype
```

```python
df['Order Item Quantity'] = pd.to_numeric(df['Order Item Quantity'] , errors ='coerce')
```

```python
df['Order Item Quantity'].dtype
```

```python
df['order date (DateOrders)'].dtype
```

```python
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'] , errors='coerce')
```

```python
df['order date (DateOrders)'].dtype
```

```python
df['shipping date (DateOrders)'].dtype
```

```python
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'] , errors='coerce')
```

```python
df['shipping date (DateOrders)'].dtype
```

Handling Missing Values

```python
df['Sales'].fillna(df['Sales'].median() )
df['Order Item Quantity'].fillna(df['Order Item Quantity'].median )
```
