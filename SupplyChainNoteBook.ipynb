Handling Missing Values

```python
df['Sales'].fillna(df['Sales'].median() )
df['Order Item Quantity'].fillna(df['Order Item Quantity'].median )
```

```python
df['Customer City'].fillna(df['Customer City'].mode()[0] , inplace =True)
df['Customer State'].fillna(df['Customer State'].mode()[0] , inplace= True)
```

Removing Duplicates 

```python
df= df.drop_duplicates()
```

Fixing Inconsistent Data 

```python
# 1. Use .loc to modify the column safely
df.loc[:, "Customer Country"] = df["Customer Country"].str.strip().str.upper()

# 2. Remove inplace=True and assign back using .loc
df.loc[:, "Customer Country"] = df["Customer Country"].replace({'IND': "INDIA"})

# 3. Repeat for other columns
df.loc[:, "Order Status"] = df["Order Status"].str.upper()
df.loc[:, "Shipping Mode"] = df["Shipping Mode"].str.upper()
```

Handling Negative Values 

```python
df = df[df['Sales'] >= 0]
df = df[df['Order Item Quantity'] >= 0]
```

Handling Outliers 

```python
df = df[df["Sales"] < df["Sales"].quantile(0.99)]
```

Adding A New Feature 

```python
df['TotalAmount'] =df['Sales'] * df['Order Item Quantity']
```

```python
df['Delivery_Time'] = (df['shipping date (DateOrders)'] - df['order date (DateOrders)']).dt.days
```

```python
df["Is_Delayed"] = df["Delivery_Time"].apply(lambda x:1 if x>5 else 0)
```

```python
df['Full_Address'] = df['Customer City'] + ','+ df['Customer State']
```

```python
import unicodedata

def clean_country_names(text):
    if not isinstance(text, str):
        return text
    
    # 1. Fix the double-encoding mess (the weird symbols in your image)
    # These often represent accented letters like Ã­, Ã¡, or Ã±
    text = text.replace('ÃÆÃ-', 'A').replace('ÃÆÃÂº', 'U').replace('ÃÆÃÂ±', 'N').replace('ÃÆÃÂ³', 'O')
    
    # 2. Standardize to ASCII (Removes remaining accents like 'Ã¼' -> 'u')
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')
    
    # 3. Clean up whitespace and case
    return text.upper().strip()

# Apply the cleaning function first
df['Order Country'] = df['Order Country'].apply(clean_country_names)
```
